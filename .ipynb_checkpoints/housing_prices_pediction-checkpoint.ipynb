## Cleaning data


```python
# checking all categorical features
for i in df.select_dtypes('object').nunique().index:
    print(i,df[i].unique())
```

###  the following categories have 'None' in them
-- bathrooms.  has 'None'  
-- sqr_feet.    has 'None'  
-- garage.      has 'None'  
--fireplaces  has empty values  
-- HOA has 'None'

```python
# replace 'None' with None so it will show when checking for missing values
df['bathrooms']=df['bathrooms'].replace('None',None).astype(float)
df['garage']=df['garage'].replace('None',None).astype(float)
df['fireplaces']=df['fireplaces'].replace(' ',None).astype(float)
df['HOA']= df['HOA'].str.replace(',','').replace('None', None).astype(float)
df['sqrt_ft']=df['sqrt_ft'].replace('None',None).astype(float)
```

```python
# drop the columns with high cardinality
df.drop(columns=['kitchen_features', 'floor_covering', 'MLS'], inplace=True)
```

```python
#round the decimal values in number of bathroom
df.bathrooms=df.bathrooms.replace({2.5:3, 3.5:4, 4.5:5})
# round the decimal values in number of garages
df.garage=df.garage.replace({2.5:3, 3.5:4, 4.5:5})
```

```python
# fill NaN with median since HOA column is skewed
df['HOA']=df['HOA'].fillna(df['HOA'].median())
```

```python
# drop NaN values
df.dropna(inplace=True)
df.info()
```

```python
# check if there is any colinearity 
features=df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(features,annot=True)
plt.savefig('corr.png');
```

```python
# drop high multicolineaerity
df.drop(columns=['zipcode', 'bedrooms', 'bathrooms'], inplace=True)
# drop columns with least correlation and leaky comlumns
df.drop(['lot_acres', 'taxes', 'year_built'], axis=1, inplace=True)
df.info()
```

```python
model=KNNclassifier()
training_score={}
validation_score={}
start=time.time()
for k in range(5,150):
    model.fit(X_train, y_train)
    y_hat=model.predict(X_train, k)
    score=accuracy(y_train, y_hat)
    training_score[k]=round(score,3)
    y_hat_val=model.predict(X_val, k)
    scoreVal=accuracy(y_val, y_hat_val)
    validation_score[k]=round(scoreVal,3)
finish=time.time()
print(finish-start, 'seconds')
print(training_score,'\n', validation_score)
```

```python
def mae(y, y_hat):
    total=np.sum(abs(y-y_hat))
    return (total/len(y))
```

```python
y_mean= y.mean()
y_pred_baseline= [y_mean]*len(y)
print("Mean apt price:", round(y_mean,2))
mae_baseline= mae(y, y_pred_baseline)
print("Baseline MAE:", round(mae_baseline,2))
```
